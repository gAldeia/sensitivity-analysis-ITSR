```python
import numpy              as np
import matplotlib.pyplot  as plt
import itea               as sr
import matplotlib.patches as patches

from sklearn.datasets import load_diabetes
from sklearn          import linear_model, datasets
from matplotlib.path  import Path
from matplotlib.lines import Line2D
from matplotlib       import rc

plt.style.use('default')
rc('text', usetex=True)

print('Done')
```

```python
# Definindo funções auxiliares que usaremos para fazer o plot de coordenadas paralelas

def parallel_plot(host, data, names, groups, legends=False):
    
    # Menores e maiores valores de cada coluna
    datamins = data.min(axis=0)
    datamaxs = data.max(axis=0)
    
    # range de cada coluna
    datadelta = datamaxs - datamins
    
    datamins -= datadelta * 0.05  # add 5% padding below and above
    datamaxs += datadelta * 0.05

    # Para inverter uma coluna, fazer como a linha abaixo com o número da coluna
    #datamaxs[1], datamins[1] = datamaxs[1], datamins[1] 
    
    datadelta = datamaxs - datamins

    # normalizando os intervalos das colunas para serem todos no mesmo valor
    zs        = np.zeros_like(data)
    zs[:, 0]  = data[:, 0]
    zs[:, 1:] = (data[:, 1:] - datamins[1:]) / datadelta[1:] * datadelta[0] + datamins[0]

    # Criando os eixos paralelos
    axes = [host] + [host.twinx() for i in range(data.shape[1] - 1)]
    
    for i, ax in enumerate(axes):
        if datamins[i] == datamaxs[i]:
            ax.set_ylim(datamins[i]*0.5, datamaxs[i]*1.5)
        else:
            ax.set_ylim(datamins[i], datamaxs[i])
            
        ax.spines['top'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        
        if ax != host:
            ax.spines['left'].set_visible(False)
            ax.yaxis.set_ticks_position('right')
            ax.spines["right"].set_position(("axes", i / (data.shape[1] - 1)))

    # Ajustando o eixo principal (axs passado como 'host')
    host.set_xlim(0, data.shape[1] - 1)
    host.set_xticks(range(data.shape[1]))
    host.set_xticklabels(names, fontsize=14)
    host.tick_params(axis='x', which='major', pad=7)
    host.spines['right'].set_visible(False)
    host.xaxis.tick_bottom()
    
    legend_handles = [
        Line2D([0,1],[0,1], lw=lw, c=c, alpha=a, ls=ls)
        for _, c, a, ls, lw, label in groups
    ]

    
    for (init, end), color, alpha, linestyle, linewidth, label in groups:

        for j in range(init, end, 1):
            # curvas bezier
            verts = list(zip(
                [x for x in np.linspace(0, len(data[0]) - 1, len(data[0]) * 3 - 2, endpoint=True)],
                np.repeat(zs[j, :], 3)[1:-1]
            ))
            
            # path.LINETO, path.CURVE4
            codes = [Path.MOVETO] + [Path.CURVE4 for _ in range(len(verts) - 1)]
            path  = Path(verts, codes)
            patch = patches.PathPatch(
                path, facecolor='none', lw=linewidth, alpha=alpha, edgecolor=color, linestyle=linestyle,
                zorder=len(groups) - j
            )
            host.add_patch(patch)
        
    if legends:
        host.legend(
            legend_handles, [label for _, _, _, _, _, label in groups],
            loc='lower center', bbox_to_anchor=(0.5, -0.35),
            ncol=len(groups), fancybox=True, shadow=True, edgecolor=None
        )
    
    return host

def metavars(axs, bestsol, X, y, ynames):
    # Ordenando pelo coeficiente
    terms = sorted(zip(bestsol.terms, bestsol.funcs, bestsol.coeffs), key=lambda tup: tup[2], reverse=True)

    # cada termo será um plot
    for (t, f, c), ax in zip(terms, axs):

        data     = []
        groups   = []

        mask = np.ones_like(t)
        mask[t==0] = 0
            
        std = np.std(np.prod(X**t, axis=1))
        
        # pegar cada linha
        for i, x in enumerate(X):
            
            # ignorar outliers maiores que 2std 
            if np.abs(np.prod(x**t)) < 2*std:
                # adicionando a variável intermediária  
                data.append(np.append(x[mask>0], np.prod(x**t)))
            
        groups.append(((0, len(data)-1), 'y', 0.25, '-', 1.5, ''))
        ax.set_title(f'Variáveis intermediárias - termo {t}', fontsize=14, pad=12)

        data = np.array(data)

        parallel_plot(
            ax, data, np.append(ynames[mask>0], 'metavar'),
            groups, legends=False
        )
```

# Investigando as metavariáveis

- ideia de que a interação cria variáveis intermediárias novas e efetivas (avaliador)
- analisar a IT como uma transformação no espaço de variáveis (Fabrício)
- metavariável: combinação ou transformação de variáveis para criação de novas (Referência PGC)
- será que é possível atribuir algum significado para essas variáveis intermediárias à partir dos significados das variáveis originais?

(inspirados agora no site do assistente de laboratório): Particularmente, analisar o IT como uma transformação no espaço de
variáveis me parece interessante e pertinente, você fez algo minimamente nessa linha no seu site, quando você plota cada termo em função da variável-alvo.

Primeiro, vamos fazer plots de coordanadas paralelas para cada variável intermediária (ou seja, cada função de interação). Um argumento a favor dessas variáveis é que estamos vendo uma função que passou por um "processo evolutivo", então ela realmente deve ser efetiva. Nessa discussão é possível entrar no mérito do expolim, e como ele define o espaço de variáveis para a transformação. Talvez, na análise de sensibilidade, maxterms tenha sido determinante para reduzir o erro (RMSE) (e, colateralmente, aumentar a interpretabilidade ao controlar o tamanho da expressão), mas a análise do expolim controlando o espaço das variáveis pode ser interessante. Lembrando que a ideia da IT é limitar o espaço de busca no conjunto de funções matemáticas, o expolim pode ser um limitante no espaço de variáveis na criação de intermediárias.

Acho que podemos ver isso de vários jeitos - ver a porcentagem de vezes que cada variável intermediária (interação) aparece na melhor solução ou população, ver a ocorrência de cada função de transformação para cada variável intermediária, ver como cada uma se relaciona com as originais (proporcional, inversamente proporcional), quando elas tendem a zero, etc - preciso pensar com mais calma nisso.

```python
X, y, coef = datasets.make_regression(n_samples=200, n_features=4,
                                      n_informative=2, noise=10,
                                      coef=True, random_state=0)

ynames     = np.array([f'x{i}' for i, _ in enumerate(X[0])])

# Vamos mudar um pouco a configuração - outro modelo, intervalo de expolim, menos funções, e mais termos.
params = {
    'popsize'  : 100,
    'gens'     : 100, 
    'minterms' : 1,
    'model'    : linear_model.LinearRegression(n_jobs=-1),
    'funs'     : {
        "id"      : lambda x: x,
        "sin"     : np.sin, 
        "cos"     : np.cos,        
        "tanh"    : np.tanh,
        "sqrt.abs": lambda x: np.sqrt(np.absolute(x))
    },
    'expolim'  : (-2, 2),
    'maxterms' : 5,
}

itea = sr.ITEA(**params)
bestsol = itea.run(X, y, log=None, verbose=True)

print('----')
print(bestsol)
```

```python
fig, axs = plt.subplots(nrows=bestsol.len, ncols=1, figsize=(10,5*bestsol.len))

metavars(axs, bestsol, X, y, ynames)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
